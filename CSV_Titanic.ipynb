```python
import csv
import time
# Função retornando a média de idade dos passageiros
def mediaIdadePassengers(dados):
  sum = 0
  for x in range(1,len(dados)):
    if(dados[x][5] != 'NaN' and dados[x][5]!= ''):
      sum = sum + (float(dados[x][5]))
  mediaAge = sum/len(dados)
  return print(f'Media de idade dos passageiros - {mediaAge:,.2f}\n\n')

# Função que retorna leitura em relação aos filhos que estavam no Titanic
def readerSons(dados):
  contSonsOrphan = 0
  contFatherNotSurvider = 0
  contSons = 0
  for x in range(1,len(dados)):
   
    contSons = contSons + float(dados[x][7])

    if(dados[x][4] == 'female' and dados[x][1] == '0'):
      contSonsOrphan = contSonsOrphan + float(dados[x][7])
    elif(dados[x][4] == 'male' and dados[x][1] == '0'):
      contFatherNotSurvider = contFatherNotSurvider + float(dados[x][7])

  return print(f'Quantidade de filhos que perderam o Pai - {contFatherNotSurvider}\nQuantidade filhos que perderam a Mãe - {contSonsOrphan}\nQuantidade de filhos embarcados - {contSons}\n\n')
      
# Função retornando a média de tarifas dos passageiros
def mediaFare(dados):
  sum = 0
  for x in range(1,len(dados)):
      sum = sum + (float(dados[x][9]))
  mediaFare = sum/len(dados)
  return print(f'Media de tarifa dos passageiros - {mediaFare:,.2f}\n\n')

# Contador de sobreviventes
def countSurviver(dados):
  contSurviver = 0    
  for x in range(1,len(dados)):
    if (dados[x][1] == '1'):
      contSurviver += 1
  return (contSurviver)

# Contador de sobreviventes dividindo por sexo
def countSurviverSex(dados):  
  contSurviverMale = 0    
  contSurviverFamale = 0    
  for x in range(1,len(dados)):
    if (dados[x][1] == '1' and dados[x][4] == 'male'):
      contSurviverMale += 1
    elif(dados[x][1] == '1' and dados[x][4] == 'female'):
      contSurviverFamale += 1

  return print(f'Sobreviventes Homem - {contSurviverMale}\nSobreviventes Mulher - {contSurviverFamale}\n\nÉ POSSÍVEL OBSERVAR QUE A QUANTIDADE DE MULHERES SOBREVIVENTES É BEM SUPERIOR A DE HOMENS, ISSO COLOCA EM PROVA A POLÍTICA DE DESEMBARQUE ADOTADA NO MOMENTO DA TRAGÉDIA.\n\n')

with open("/content/drive/MyDrive/titanic.csv",'r') as arquivo:
    leitor = csv.reader(arquivo)
    dados = list(leitor)
test = 1    
while(test == 1):
  option = str(input('1 - Total de sobreviventes\n\n2 - Sobreviventes separados por sexo\n\n3 - Media de idade dos passageiros\n\n4 - Media de valor das tarifas\n\n5 - Quantidade de filhos a bordo do Titanic\n\n6 - Finalizar\n\n'))
  print('\n')
  if(option == '1'):
    print(f'Quantidade total de sobreviventes - {countSurviver(dados)}\n\n')
  elif(option == '2'):
    countSurviverSex(dados)
  elif(option == '3'):
    mediaIdadePassengers(dados)
  elif(option == '4'):
    mediaFare(dados)
  elif(option == '5'):
    readerSons(dados)
  elif(option == '6'):
    test = 0
    print('Finalizando...')
    time.sleep(1)
    print('-')
    time.sleep(1)
    print('-')
    time.sleep(1)
    print('-')
    time.sleep(1)
    print('-')
    time.sleep(1)
    print('Sistema encerrado. Obrigado !')

```
